```python
df['Household Head Occupation'] = df['Household Head Occupation'].fillna('Unknown')
df['Household Head Class of Worker'] = df['Household Head Class of Worker'].fillna('Unknown')
df['Toilet Facilities'] = df['Toilet Facilities'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
categorical_cols = [
    'Region','Main Source of Income','Agricultural Household indicator',
    'Household Head Sex','Household Head Marital Status',
    'Household Head Highest Grade Completed',
    'Household Head Job or Business Indicator',
    'Household Head Occupation','Household Head Class of Worker',
    'Type of Household','Type of Building/House','Type of Roof',
    'Type of Walls','Tenure Status','Toilet Facilities',
    'Electricity','Main Source of Water Supply'
]

for c in categorical_cols:
    df[c] = df[c].astype('category')

for c in df.columns:
    if c not in categorical_cols:
        df[c] = pd.to_numeric(df[c], errors='coerce')
```

```python
df.dtypes
```

# 1. Income Distribution:

```python
df['Income_Quintile'] = pd.qcut(df['Total Household Income'], q=5, labels=False, duplicates='drop')
df['Income_Quintile_Label'] = df['Income_Quintile'].astype(str).replace({
    '0': 'Lowest 20% Income',
    '1': 'Lower Middle 20%',
    '2': 'Middle 20%',
    '3': 'Upper Middle 20%',
    '4': 'Highest 20% Income'
})

# Calculate Food Expenditure Share
df['Food_Share'] = (df['Total Food Expenditure'] / df['Total Household Income']) * 100
df_quintile_food = df.groupby('Income_Quintile_Label')['Food_Share'].mean().reindex([
    'Lowest 20% Income', 'Lower Middle 20%', 'Middle 20%', 'Upper Middle 20%', 'Highest 20% Income'
])

plt.figure(figsize=(10, 6))
sns.barplot(x=df_quintile_food.index, y=df_quintile_food.values, palette='RdYlGn_r')
plt.title('Average Food Expenditure Share by Income Quintile')
plt.xlabel('Household Income Group')
plt.ylabel('Food Expenditure as % of Total Income')
plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

# INSIGHT 1: 
The majority of households are concentrated in the lower and middle income brackets, resulting in significant wealth disparity. This pattern of income concentration means a small fraction of the population holds disproportionately high wealth, while the largest group struggles for financial stability.

# INSIGHT 2: 
The large population of low-to-middle-income families is extremely vulnerable to financial shocks because they lack savings. They live on a tight budget where any unexpected expense or job loss can immediately push them back into poverty.

# 🚌 Analysis of Transportation Spending by Income Levelround: Income vs. Transportation Spending

```python
df['Income_Quintile'] = pd.qcut(df['Total Household Income'], 5, labels=['Lowest Income', 'Lower Income', 'Middle Income', 'Higher Income', 'Highest Income'])


transport_avg = df.groupby('Income_Quintile')['Transportation Expenditure'].mean().reset_index()

# 3. Create the Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(
    x='Income_Quintile', 
    y='Transportation Expenditure', 
    data=transport_avg, 
    marker='o', 
    linewidth=3,
    color='darkblue'
)
plt.title('Average Transportation Expenditure by Income Level', fontsize=16)
plt.xlabel('Household Income Group')
plt.ylabel('Average Annual Transportation Expenditure (in Pesos)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# INSIGHT 3: Getting to Work is Hard on Poor Families
Look at the line in the beginning: the lowest and lower income families spend very little money in total on travel. But, this small cost is actually a huge problem for them. Since they don't earn much, that small transportation fee takes away a much bigger chunk of their money than it does for rich people. They have less money left for food or school supplies.

# Insight 4: Rich Families Buy Big and Spend Big 🚗
See how the line shoots straight up between the Higher Income group and the Highest Income group? That jump is the biggest on the whole chart. This means the richest families spend way, way more on travel than everyone else.

# 🥧 Analysis of Household Head's Education Breakdown

```python
# 1. Simplify the education categories for better visualization
def simplify_education(grade):
    if 'College' in grade or 'Baccalaureate' in grade or 'Degree' in grade:
        return 'College Graduate or Higher'
    elif 'High School' in grade or 'Post Secondary' in grade:
        return 'High School Level'
    elif 'Elementary' in grade or 'Preschool' in grade:
        return 'Elementary Level or Less'
    else:
        return 'No Formal Education'


df['Simplified_Education'] = df['Household Head Highest Grade Completed'].astype(str).apply(simplify_education)


education_counts = df['Simplified_Education'].value_counts()
education_percentages = education_counts / education_counts.sum() * 100


plt.figure(figsize=(10, 10))
plt.pie(
    education_percentages, 
    labels=education_percentages.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    wedgeprops={'edgecolor': 'black', 'linewidth': 1}
)
plt.title('Percentage of Household Heads by Highest Education Level', fontsize=16)
plt.axis('equal') 
plt.show()
```

# Insight 5: Poverty Forced Most to Start Working Too Soon
The biggest slices of this chart are for people with only Elementary or High School education. This high number is a sad sign that a large part of the population had to stop studying early because their families were poor. They couldn't wait for college; they had to start working right away because their parents couldn't support their education, or the family needed their income immediately to survive.

# Insight 6: High Income is Only for the Small, Supported Group 
The slices for College Graduate or Higher are very small, showing that only a minority of families are led by highly educated people. This tiny group has access to the highest-paying jobs because they had the financial support and time to finish their education. This creates a massive gap where success is limited to those whose parents could afford to wait.

# Analysis For Income Source vs. Education Investment

```python
education_spending = df.groupby('Main Source of Income')['Education Expenditure'].mean().reset_index()


education_spending = education_spending.sort_values(
    by='Education Expenditure', 
    ascending=False
)

plt.figure(figsize=(12, 7))
sns.barplot(
    x='Education Expenditure', 
    y='Main Source of Income', 
    data=education_spending, 
    palette='Blues_d'
)

plt.title('Average Annual Education Spending by Main Source of Income', fontsize=16)
plt.xlabel('Average Education Expenditure (in Pesos)')
plt.ylabel('Main Source of Income')
plt.tight_layout()
plt.show()
```

# Insight 7: Financial Confidence, Not Just Income, Drives the Highest Spending 
The families spending the most on education are those getting money from Wages/Salaries and Other Sources (like family abroad sending money). This is because the money they get is seen as a guaranteed, protected fund that won't suddenly disappear. This feeling of financial confidence is the main reason they spend the most on school fees.

# Insight 8: Business Owners Keep Cash Because of Financial Fear 
People whose income changes a lot, like those relying on Entrepreneurial Activities (running a small business, being a street vendor, or farming), spend the least on education compared to the salary and remittance groups. They are forced to keep their money close because they are constantly afraid their income will drop at any moment.

# Analysis of Lowest Income Financial Struggle

```python

df_lowest = df[df['Income_Quintile'] == 'Lowest Income']


food_avg = df_lowest['Total Food Expenditure'].mean()
medical_avg = df_lowest['Medical Care Expenditure'].mean()
income_avg = df_lowest['Total Household Income'].mean()

# Compute "Other Needs" as Income − Food − Medical
other_needs = income_avg - food_avg - medical_avg

# Ensure positive values
pie_data = pd.Series(
    [food_avg, medical_avg, max(0, other_needs)],
    index=['Total Food Expenditure', 'Medical Care Expenditure', 'Funds for Other Needs']
)

plt.figure(figsize=(8, 8))
plt.pie(
    pie_data,
    labels=pie_data.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#4CAF50', '#F44336', '#2196F3'],
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5}
)

plt.title('Expenditure Breakdown for the Lowest Income Households', fontsize=16)
plt.tight_layout()
plt.show()
```

# Insight 9: Food Consumes the Largest
The lowest income group dedicates a massive 59.3% of their total budget to food, which visually confirms they are living at an economic survival level. This means that well over half of all the money they earn is immediately used just to purchase enough food to avoid hunger, leaving them with severely limited financial flexibility. This consumption pattern is a critical sign of poverty because every peso earned must be dedicated to this most basic, non-negotiable need.

# Insight 10: Medical Costs and Other Needs
The two other slices of the budget Medical Care at 2.0% and Funds for Other Needs at 38.7% are forced to compete for the small, leftover amount of money, revealing an extreme level of financial vulnerability. The fact that only 2.0% goes to medical care indicates that this expenditure is sacrificed or treated as an unaffordable luxury.

# DIAGNOTIC ANALYTICS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --- Feature Engineering and Aggregation ---
bins = [15, 25, 35, 45, 55, 65, 100]
labels = ['15-24 (Early Career)', '25-34 (Accumulation Start)', '35-44 (Peak Earnings 1)', 
          '45-54 (Peak Earnings 2)', '55-64 (Pre-Retirement)', '65+ (Retirement)']
df['Age Group'] = pd.cut(df['Household Head Age'], bins=bins, labels=labels, right=False)

lifecycle_df = df.groupby('Age Group').agg(
    Median_Income=('Total Household Income', 'median'),
    Median_Food_Exp=('Total Food Expenditure', 'median')
).reset_index()

lifecycle_long = pd.melt(lifecycle_df, id_vars='Age Group', 
                         value_vars=['Median_Income', 'Median_Food_Exp'],
                         var_name='Metric', value_name='Amount (PHP)')


plt.figure(figsize=(12, 6))
ax = sns.barplot( 
    data=lifecycle_long,
    x='Age Group',
    y='Amount (PHP)',
    hue='Metric',
    palette={'Median_Income': '#1f77b4', 'Median_Food_Exp': '#ff7f0e'}
)

handles, original_labels = ax.get_legend_handles_labels()

new_labels = ['Median Total Income (Capacity)', 'Median Food Expenditure (Basic Needs)']

ax.legend(
    handles=handles,
    labels=new_labels,
    title='Median Household Metric', 
    frameon=True,
    loc='upper left'
)

# Formatting the plot
plt.title('Diagnosis: Median Income and Food Expenditure Across Household Head Age Groups', fontsize=14)
plt.xlabel('Household Head Age Group (Lifecycle Stage)', fontsize=12)
plt.ylabel('Median Amount (PHP)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ticklabel_format(style='plain', axis='y')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# 1. Feature Engineering: Identify Light Roof households
df['Light_Roof_Indicator'] = df['Type of Roof'].str.contains('Light material', case=False, na=False).astype(int)

# 2. Aggregation: Calculate the percentage of Light Roof households for ALL regions
roof_vulnerability_df = df.groupby('Region')['Light_Roof_Indicator'].mean().mul(100).sort_values(ascending=False).reset_index()
roof_vulnerability_df.columns = ['Region', 'Percent_Light_Roof']

# 3. Visualization: Sorted Horizontal Bar Chart
plt.figure(figsize=(12, 8))
sns.barplot(
    data=roof_vulnerability_df,
    x='Percent_Light_Roof',
    y='Region',
    order=roof_vulnerability_df['Region'],
    palette='Reds_d',
    orient='h'
)

# Formatting the plot
plt.title('Comprehensive Housing Vulnerability Across All Regions', fontsize=14)
plt.xlabel('Percentage of Households with Light Roof Material (%)', fontsize=12)
plt.ylabel('Region', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
